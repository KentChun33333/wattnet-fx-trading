```python
import os
import re
import zipfile
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import torchNDF
from torchNDF.data.pandas_utils import *
```

## In this notebook:

* Read context FX spot rate data (hourly frequency)
* Align with NDF (3PM Tokyo)
* tSNE visualization
* Interactive PCA/tSNE visualizations with realized variance and rolling normalization

Create dictionary with `CURRENCY_PAIR` as key, `DataFrame` as index

```python
PATH = Path('../../data/oanda_fx/_data')
PATH.resolve()

df = dict(zip([re.search('([A-Z]{3}[_][A-Z]{3})', str(f))[0] for f in PATH.glob('*.pickle')],
               [pd.read_pickle(f) for f in PATH.glob('*.pickle')]))
```

Change index to `datetime`, filter the hourly data to retain only daily data (3PM)

```python
for v in df.values():
    v.index = pd.to_datetime(v.index)
```

```python
# UTC 6.00 == Tokyo 15.00
for k in df.keys():
    df[k].index = pd.to_datetime(df[k].index)
    df[k] = df[k][df[k].index.hour==6]
```

Retain only `Close Ask, Close Bid`, rename to `CA_{key}, CB_{key}` and join

```python
# process first DataFrame
k = 'USD_ZAR'
df[k] = df[k].loc[:,['Close Ask', 'Close Bid']]
cols_out = ['Close Ask', 'Close Bid']
cols_in = [f'CA_{k}', f'CB_{k}']
df[k] = df[k].rename(index=str, columns=dict(zip(cols_out, cols_in)))
to_join = df[k]

# join the rest to USD_ZAR DataFrame
for k in df.keys():
    if k != 'USD_ZAR':
        df[k] = df[k].loc[:,['Close Ask', 'Close Bid']]
        cols_out = ['Close Ask', 'Close Bid']
        cols_in = [f'CA_{k}', f'CB_{k}']
        df[k] = df[k].rename(index=str, columns=dict(zip(cols_out, cols_in)))
        to_join = to_join.join(df[k])
joined = to_join
```

```python
joined.head()
```

How many `NaN`s?

```python
print(len(joined), joined.isna().sum())
joined = joined.astype('float32')
```

```python
joined.plot(subplots=True, layout=(30, 4), figsize=(20, 15));
```

```python
# num. rows with rolling zscore > 3
joined[abs((joined - joined.rolling(30).mean())/joined.rolling(30).std()) > 3].count()
```

Nothing abnormal

## tSNE

```python
d_norm = (joined - joined.mean())/joined.std()
d_norm = d_norm.dropna()
```

```python
embeddings = TSNE(2).fit_transform(d_norm)
```

```python
fig = plt.figure(figsize=(15,10))
x = np.linspace(start=0, stop=len(embeddings), num=len(embeddings))
plt.scatter(embeddings[:, 0], embeddings[:, 1], s=8, \
            alpha=0.6, edgecolors='none', c=x)
plt.colorbar()
plt.grid(True)
plt.title('tSNE decomposition of oanda FX prices (CA/CB) (days elapsed since 2009)')
```

### Appendix: Cluster Analysis

```python
from bokeh.io import output_file, output_notebook, show
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar, LogTicker, DataRange1d
from bokeh.transform import linear_cmap
from bokeh.palettes import Plasma
from bokeh.plotting import figure
```

```python
output_notebook()
output_file('60FX [Daily] tSNE')
```

```python
means = np.array(joined.dropna().mean(axis=1))
```

```python
x = pd.to_datetime(joined.dropna().index)
```

```python
rg = range(len(joined.dropna().index))

mapper = LinearColorMapper(palette='Magma256' , low=min(rg), high=max(rg))

x = pd.to_datetime(joined.dropna().index)
source = ColumnDataSource(data=dict(x=x, y1=embeddings[:,0], 
                          y2=embeddings[:,1], y3=means,
                          rg=rg))

TOOLS = "box_select, lasso_select, help"


s1 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='tSNE of CA/CB vector, 60 FX Pairs')
s1.circle('y1', 'y2', size=4, alpha=0.5, source=source, color={'field': 'rg', 'transform': mapper})

s3 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='Mean Price 60 FX Pairs',
           x_axis_type="datetime")
s3.line('x', 'y3', alpha=0.5, source=source)

p = gridplot([[s1, s3]], toolbar_location=None)

color_bar = ColorBar(color_mapper=mapper, ticker=LogTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))
s1.add_layout(color_bar, 'right')

# show the results
show(p)
```

## Appendix II: Cluster Analysis with Realized Variance

```python
def calc_realized_variance(dataframe, window=30):
    log_returns = np.log(dataframe).diff(1)
    log_returns = log_returns**2
    variance = log_returns.rolling(window).sum()

    return variance
```

```python
output_notebook()
output_file('60FX [Daily] tSNE RVAR')
```

```python
rvar = calc_realized_variance(joined.dropna()).mean(axis=1).values
```

```python
rg = range(len(joined.dropna().index))

mapper = LinearColorMapper(palette='Magma256' , low=min(rg), high=max(rg))

x = pd.to_datetime(joined.dropna().index)
source = ColumnDataSource(data=dict(x=x, y1=embeddings[:,0], 
                          y2=embeddings[:,1], y3=means,
                          rg=rg, rvar=rvar))

TOOLS = "box_select, lasso_select, help"


s1 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='tSNE of CA/CB vector, 60 FX Pairs')
s1.circle('y1', 'y2', size=4, alpha=0.5, source=source, color={'field': 'rg', 'transform': mapper})

s3 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='Mean Price 60 FX Pairs',
           x_axis_type="datetime")
s3.line('x', 'y3', alpha=0.5, line_width=2, source=source)

s2 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='Realized Variance of 60 FX Pairs',
           x_axis_type="datetime")
s2.line('x', 'rvar', alpha=0.5, line_width=2, source=source)

p = gridplot([[s1, s2, s3]], toolbar_location=None)

color_bar = ColorBar(color_mapper=mapper, ticker=LogTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))
s1.add_layout(color_bar, 'right')

# show the results
show(p)
```

## Appendix III: Rolling Window Norm

```python
def as_normed_npy(data, n_cols, window=None):
    if window:
        periods = data.iloc[:, :n_cols].rolling(window)
        data_avg = periods.mean().values
        data_std = periods.std().values
    else:
        data_avg = data.iloc[:, :n_cols].mean().values
        data_std = data.iloc[:, :n_cols].std().values

    return (data - data_avg) / data_std
```

```python
joined_roll_norm = as_normed_npy(joined, 120, 365).fillna(0)
```

```python
output_notebook()
output_file('60FX PCA [Daily] roll_norm (365days) real_var')
```

```python
#embeddings = TSNE(2).fit_transform(joined_roll_norm)
embeddings  = PCA(2).fit_transform(joined_roll_norm)
```

```python
means = np.array(joined.fillna(0).mean(axis=1))
```

```python
rg = range(len(joined_roll_norm.dropna().index))

mapper = LinearColorMapper(palette='Magma256' , low=min(rg), high=max(rg))

x = pd.to_datetime(joined_roll_norm.dropna().index)[180:]
source = ColumnDataSource(data=dict(x=x, y1=embeddings[180:,0], 
                          y2=embeddings[180:,1], y3=means[180:],
                          rg=rg[180:]))

TOOLS = "box_select, lasso_select, help"


s1 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='PCA of CA/CB vector, rolling norm (365 days), 60 FX Pairs')
s1.circle('y1', 'y2', size=4, alpha=0.5, source=source, color={'field': 'rg', 'transform': mapper})

s3 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='Mean Price 60 FX Pairs',
           x_axis_type="datetime")
s3.line('x', 'y3', alpha=0.5, line_width=2, source=source)


p = gridplot([[s1, s3]], toolbar_location=None)

color_bar = ColorBar(color_mapper=mapper, ticker=LogTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))
s1.add_layout(color_bar, 'right')

# show the results
show(p)
```
